```
# Creating a parent class Person
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Creating child class Student
class Student(Person):
    def __init__(self, name, age, student_id):
        # Using super() to initialize the inherited attributes
        super().__init__(name, age)
        self.student_id = student_id

    def enroll_course(self, course_name):
        print(f"{self.name} has enrolled in the course: {course_name}.")


    def display_info(self):
        super().display_info()
        print(f"Student ID: {self.student_id}")

#  Creating child class Lecturer
class Lecturer(Person):
    def __init__(self, name, age, employee_id):

        super().__init__(name, age)
        self.employee_id = employee_id

    def teach_course(self, course_name):
        print(f"Lecturer {self.name} is teaching the course: {course_name}.")



# number 4

if __name__ == "__main__":

    student1 = Student("Kwame", 21, "UMaT-STU-001")
    lecturer1 = Lecturer("Dr. Mensah", 45, "UMaT-EMP-105")

    print("=== Student Demonstration ===")

    student1.display_info()

    student1.enroll_course("Introduction to OOP")

    print("\n=== Lecturer Demonstration ===")

    lecturer1.display_info()

    lecturer1.teach_course("Advanced Data Structures")
```
